```python
# Load your trained YOLO model
model = YOLO(r'E:\Clg Activities\Projects\Cosmetic shade recomender\Project code\best.pt')  # Replace with your model's file path
```

```python
def recommend_shades(image_path):
    # Run the model and obtain results
    results = model(image_path, conf=0.3)  # Adjust confidence threshold if needed
    predictions = results[0]  # Access first image results

    # Extract class probabilities
    skin_tone_classes = ['Very Fair', 'Fair', 'Medium', 'Medium Dark', 'Dark']

    if hasattr(predictions, 'probs') and predictions.probs is not None:
        # Convert probs to a NumPy array
        probs = np.array(predictions.probs.data)  # Ensure compatibility with NumPy
    else:
        print("No bounding boxes found. Using default probabilities.")
        probs = np.array([0.38, 0.33, 0.13, 0.09, 0.07])  # Example probabilities for testing

    # Predict the skin tone
    predicted_skin_tone_index = np.argmax(probs)
    predicted_skin_tone = skin_tone_classes[predicted_skin_tone_index]
    print(f"Predicted Skin Tone: {predicted_skin_tone}")

    # Fetch recommendations for foundation shades
    foundation_shades = {
        "L'oreal Paris Infallible 24h Fresh Wear": f_shade_LP[predicted_skin_tone],
        "Maybelline Fit Me": f_shade_M[predicted_skin_tone],
        "Sugar Ace Of Face": f_shade_S[predicted_skin_tone]
    }

    # Display foundation recommendations
    print("\nRecommended Foundation Shades:")
    for brand, shades in foundation_shades.items():
        print(f"{brand}: {shades}")

    # Visualize foundation shades in a table
    df_foundation = pd.DataFrame([
        [brand, shade]
        for brand, shades in foundation_shades.items()
        for shade in shades
    ], columns=['Name', 'Foundation Shade Code'])

    fig, ax = plt.subplots(figsize=(6, len(df_foundation) // 2))
    ax.axis('off')
    table = plt.table(cellText=df_foundation.values, colLabels=df_foundation.columns, loc='center', cellLoc='center')
    table.auto_set_font_size(False)
    table.set_fontsize(10)
    table.auto_set_column_width([0, 1])
    plt.title("Foundation Shade Recommendations", pad=20)
    plt.show()

    # Fetch recommendations for concealer shades
    concealer_shades = {
        "L'oreal Paris Infallible Full Wear": f_shade_LP_C[predicted_skin_tone],
        "Maybelline Fit Me": f_shade_M_C[predicted_skin_tone],
        "Sugar Magic Wand Waterproof": f_shade_S_C[predicted_skin_tone]
    }

    # Display concealer recommendations
    print("\nRecommended Concealer Shades:")
    for brand, shades in concealer_shades.items():
        print(f"{brand}: {shades}")

    # Visualize concealer shades in a table
    df_concealer = pd.DataFrame([
        [brand, shade]
        for brand, shades in concealer_shades.items()
        for shade in shades
    ], columns=['Name', 'Concealer Shade Code'])

    fig, ax = plt.subplots(figsize=(6, len(df_concealer) // 2))
    ax.axis('off')
    table = plt.table(cellText=df_concealer.values, colLabels=df_concealer.columns, loc='center', cellLoc='center')
    table.auto_set_font_size(False)
    table.set_fontsize(10)
    table.auto_set_column_width([0, 1])
    plt.title("Concealer Shade Recommendations", pad=20)
    plt.show()
```

```python
import cv2

def display_image(image_path):
    # Load the image using OpenCV
    image = cv2.imread(image_path)
    # Convert BGR to RGB because OpenCV loads images in BGR format
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    # Display the image
    plt.imshow(image)
    plt.axis('off')  # Turn off axis labels
    plt.show()
```
